Segmentation Basis:
Clusters will be based on patterns in reservation behaviors, such as lead time, repeated_guest status, noofprevious_cancellations, and avgpriceper_room, as well as guest preferences like typeofmeal_plan, room type reserved, and noofspecial_requests. The objective is to group customers with similar behaviors and preferences, which will aid hotels in formulating tailored marketing and service offerings.


```python
required_features = ['lead_time', 'repeated_guest', 'no_of_previous_cancellations', 'avg_price_per_room', 'type_of_meal_plan', 'room_type_reserved', 'no_of_special_requests']
```

```python
df_req = df[required_features]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

#features that need to be scaled are : lead_time, avg_price_per_room
selected_features = ['lead_time', 'avg_price_per_room']
scaler = StandardScaler()
df_req_SS = df_req.copy()
df_req_SS[selected_features] = scaler.fit_transform(df_req[selected_features])
min_max_scaler = MinMaxScaler()
df_req_MM = df_req.copy()
df_req_MM[selected_features] = min_max_scaler.fit_transform(df_req[selected_features])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt

best_silhouette_score = -1
best_davies_bouldin_score = float('inf')
optimal_num_clusters = 1
for n_clusters in range(2, 11): 
    kmeans_SS = KMeans(n_clusters = n_clusters, random_state=42)
    cluster_labels = kmeans_SS.fit_predict(df_req_SS)
    
    silhouette_avg = silhouette_score(df_req_SS, kmeans_SS.labels_)
    davies_bouldin_avg = davies_bouldin_score(df_req_SS, kmeans_SS.labels_)
    print("For n_clusters =", n_clusters,"The average silhouette_score is :", silhouette_avg)
    
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_davies_bouldin_score = davies_bouldin_avg
        optimal_num_clusters = n_clusters
```

```python
kmeans_SS = KMeans(n_clusters=optimal_num_clusters, random_state=42)
kmeans_SS.fit(df_req_SS)
df_req_SS['Cluster_Labels_km_MM'] = kmeans_SS.labels_

silhouette_score_value_km_SS = silhouette_score(df_req_SS, kmeans_SS.labels_)
davies_bouldin_score_value = davies_bouldin_score(df_req_SS, kmeans_SS.labels_)
print('optimal_num_clusters :', optimal_num_clusters)
print('silhouette_score_value kmeans using MinMaxScaler: ', silhouette_score_value_km_SS)
print('davies_bouldin_score_value: ', davies_bouldin_score_value)
```
